```python
import numpy as np
from matplotlib import pyplot as plt
from pulse_method.toolbox_pulse import *
```

```python
operator=thermograms()
```

```python
data=np.load('/Users/jaworskj/Desktop/AGH_data_sample/Open_source_data/sequences_np/Z_002.npy')
```

```python
plt.title('Example thermogram')
plt.imshow(data[250,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')
plt.colorbar()
```

```python
plt.title('Comparison of the defect and sound region time evolution')
plt.plot(data[:,180,80],label='Sound material')
plt.plot(data[:,90,170],label='Defect')
plt.legend()
plt.xlim([0,50])
plt.grid()
plt.ylabel('Thermal value')
plt.xlabel('Frames')
```

```python
result_ppt,_=operator.PPT(data)
```

```python
result_pct,_=operator.PCT(data,n_components=5)
```

```python
result_tsr=operator.TSR(data)
```

```python
result_hos=operator.HOS(data)
```

```python
result_dmd,_=operator.DMD(data,truncation=7)
```

```python
plt.figure(figsize=(15,11))
plt.subplot(2,3,1)
plt.title('Example thermogram')
plt.imshow(data[250,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,2)
plt.title('Pulse Phase Thermography')
plt.imshow(result_ppt[1,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,3)
plt.title('Principal Component Thermography')
plt.imshow(result_pct[4,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,4)
plt.title('Thermographic Signal Reconstruction')
plt.imshow(result_tsr[0,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,5)
plt.title('High Order Statistic')
plt.imshow(result_hos[0,:,:],cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,6)
plt.title('Dynamic Mode Decomposition')
plt.imshow(result_dmd[-2,:,:].imag,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')
```

```python
binary1=operator.binarize_mask(data[250,:,:],threshold_value=0.58)
binary2=operator.binarize_mask(result_ppt[1,:,:],threshold_value=0.42)
binary3=operator.binarize_mask(result_pct[4,:,:],threshold_value=0.38)
binary4=operator.binarize_mask(result_tsr[0,:,:],threshold_value=0.38)
binary5=operator.binarize_mask(result_hos[0,:,:],threshold_value=0.15)
binary6=operator.binarize_mask(result_dmd[-2,:,:].imag,threshold_value=0.6)
```

```python
plt.figure(figsize=(15,11))
plt.subplot(2,3,1)
plt.title('Example thermogram')
plt.imshow(binary1,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,2)
plt.title('Pulse Phase Thermography')
plt.imshow(binary2,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,3)
plt.title('Principal Component Thermography')
plt.imshow(binary3,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,4)
plt.title('Thermographic Signal Reconstruction')
plt.imshow(binary4,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,5)
plt.title('High Order Statistic')
plt.imshow(binary5,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')

plt.subplot(2,3,6)
plt.title('Dynamic Mode Decomposition')
plt.imshow(binary6,cmap='grey')
plt.xlabel('Width of a frame')
plt.ylabel('Height of a frame')
```
